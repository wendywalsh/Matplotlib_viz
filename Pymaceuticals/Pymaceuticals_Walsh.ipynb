```python
#Import dependencies
import pandas as pd
import matplotlib as plt
import numpy as np
import os
```

```python
#Read in csv files
filepath = os.path.join("raw_data","clinicaltrial_data.csv")
filepath2 = os.path.join("raw_data","mouse_drug_data.csv")
```

```python
clinical_rawdata_df = pd.read_csv(filepath, low_memory = False)
clinical_rawdata_df.head()
```

```python
mouse_rawdata_df = pd.read_csv(filepath2, low_memory = False)
mouse_rawdata_df.head()

#1 duplicate = g989 mouse_rawdata_df.nunique() vs count()
```

```python
#Merge both tables
merged_rawdata = pd.merge(clinical_rawdata_df,mouse_rawdata_df, on="Mouse ID")
merged_rawdata=pd.DataFrame(merged_rawdata)
merged_rawdata.head()
```

```python
#Creating a scatter plot that shows how the tumor volume changes over time for each treatment. X=timepoint y=volume
#Group by treatment, by mouse ID

grouped_tumorvol = merged_rawdata.groupby(["Drug","Timepoint"]).agg({"Tumor Volume (mm3)":"mean"})
grouped_tumorvol= pd.DataFrame(grouped_tumorvol)
grouped_tumorvol
```

```python
# Scatter plot
#list =grouped_tumorvol["Drug"]
#color = orchid, firebrik, dark salmon, cyan, plum, navy, lightskyblue, green, gold, red
plt.scatter(grouped_tumorvol["Timepoint"], grouped_tumorvol["Tumor Volume (mm3)"])
#plt.scatter(grouped["No. of rides"], grouped["Avg Fare"], s=grouped["driver_count"], c = grouped["color ref"])
```

```python
#Creating a scatter plot that shows how the number of metastatic (cancer spreading) sites 
#changes over time for each treatment. 
metastatic_change = merged_rawdata.groupby(["Drug","Timepoint"]).mean()["Metastatic Sites"]
metastatic_change=pd.DataFrame(metastatic_change)
metastatic_change
```
